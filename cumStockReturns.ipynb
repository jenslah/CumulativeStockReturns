```python
tickers = ['AAPL', 'GOOGL', 'MSFT', '^GSPC']

df = pd.DataFrame()

today = dt.now().today().strftime('%Y-%m-%d')

for ticker in tickers:
    df[ticker] = yf.Ticker(ticker).history(start='2015-09-21', end='2020-09-18').Close

# df.agg(['min', 'max', 'count']) // Teil aus describe()
df.isnull().sum()
df.describe()
```

#### Compute cumulative return

```python
cum_return = ((df.iloc[-1] - df.iloc[0]) / df.iloc[0]) * 100
cum_return
```

#### Compute cumulative return from single daily returns

```python
df_daily_r = df.pct_change()
# skip first row with NA
df_daily_r = df_daily_r[1:]
df_daily_r
```

```python
# cumprod function pandas
df_cumdaily_r = (1 + df_daily_r).cumprod() - 1
df_cumdaily_r = df_cumdaily_r.reset_index()
df_cumdaily_r
```

```python
# entire period return
df_return_entire_period = df_cumdaily_r.iloc[:, 1:].tail(1)
df_return_entire_period *100
```

```python
plt.figure()
df_cumdaily_r.plot(x='Date', figsize=(15,15))
plt.style.use('ggplot')
```
